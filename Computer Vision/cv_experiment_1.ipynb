#reading and  displaying images in diffrent formats

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
import pandas as pd
```

```python
ima=cv.imread('imag1.jpg',-1)
plt.imshow(ima)
```

```python
ima=cv.imread('imag1.jpg',1)
plt.imshow(ima)
```

```python
ima=cv.imread('imag1.jpg',0)
plt.imshow(ima)
```

```python
ima=cv.cvtColor(ima,cv.COLOR_BGR2RGB)
plt.imshow(ima)
```

```python
ima=cv.imread('imag1.jpg',1)
ima=cv.cvtColor(ima,cv.COLOR_BGR2HSV_FULL)
plt.imshow(ima)
```

```python
ima=cv.imread('imag1.jpg',1)
ima=cv.cvtColor(ima,cv.COLOR_BGRA2GRAY)
plt.imshow(ima)
```

```python
ima=cv.imread('imag1.jpg',1)
ima=cv.cvtColor(ima,cv.COLOR_BGR2HSV_FULL)
plt.imshow(ima)
plt.show()
ima=cv.cvtColor(ima,cv.COLOR_HSV2BGR_FULL)
plt.imshow(ima)
plt.show()
```
